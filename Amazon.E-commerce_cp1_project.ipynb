#### importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
am = pd.read_csv(
    r'D:\sujal\Project_dataset\Amazon_dataset\Amazon Sale Report.csv',
    low_memory=False)
```

```python
am.info()
```

### handling null values

```python
am.isnull().sum()[am.isnull().sum()>0]
```

```python
am['Courier Status'] = am['Courier Status'].fillna('Unknown')
am['currency'] = am['currency'].fillna('INR')
am['Amount'] = am['Amount'].fillna(am['Amount'].median())
am['ship-city'] = am['ship-city'].fillna('Unknown')
am['ship-state'] = am['ship-state'].fillna('Unknown')
am['ship-postal-code'] = am['ship-postal-code'].fillna(0)
am['ship-country'] = am['ship-country'].fillna('India')
```

```python
am.ASIN.value_counts()
```

### dropping unnecessary columns

```python
am = am.drop(['promotion-ids'], axis =1)
am = am.drop(['fulfilled-by'], axis = 1)
am = am.drop(['Unnamed: 22'], axis = 1)
```

```python
am = am.drop(columns=['index'])
am = am.drop(columns=['ASIN'])
```

### dropping duplicates

```python
am = am.drop_duplicates()
```

### convert to proper date

```python
am['Date'] = pd.to_datetime(am['Date'])
```

### fixing amount column

```python
am['Amount'] = pd.to_numeric(am['Amount'],errors='coerce')
```

```python
am = am.dropna(subset=['Amount'])
```

### creating month & year column for dashboard 

```python
am['Month'] = am['Date'].dt.month_name()
am['Year'] = am['Date'].dt.year
```

### saving cleaned dataset

```python
am.to_csv('clean_amazon.csv', index=False)
```
